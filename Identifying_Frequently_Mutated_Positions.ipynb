```python
###Here, I use a protein alignment to identify homologous positions in different gpcrs,
#The alignment uses its own numbering scheme, so I need to create a dictionary linking alignment number to
#sequence position number for every protein

Protein_dict={}
sequence_array=[]
old_protein="cat"#initializing to gibberish, this is useful
loop_count1=0

FASTA=open(fasta_file)##This is a protein alignment of all Class A GPCRS, it
#is used to identify which sequence positions represent the same homologous position between different receptors

for line in FASTA:
    #print(line)
    if line[0]=='>':#this is a line containing the sequence name
        nametemp=line.split('_')
        #print(nametemp)
        #print(nametemp[1])
        
        try:
            Protein_name=str(nametemp[1])
        except:
            ''
        if loop_count1>0:
            Protein_dict[old_protein]=sequence_array
        sequence_array=[]
        old_protein=str(Protein_name)
        pos_count=1
        loop_count1=loop_count1+1
        #print(Protein_name)
    elif line[0]=='&':
        Protein_dict[old_protein]=sequence_array
    else:
        for position in line:
            if position != '\n':
                if position != '.':
                    place=pos_count
                    pos_count=pos_count+1
                else:
                    place=position
                try:
                    sequence_array.append(place)
                except:
                    sequence_array=[place,]
FASTA.close()

print('Protein_dict',Protein_dict['317108157'])
```

```python
###Eventually, I want to combine mutations at shared positions. If a given alignemnt position only has 2 sequences
# and every other protein has a gap there, this will always have fewer positions that an alignment position
# where every protein has an amino acid. In other words, we only want to explore positions with a lower frequency
#of gaps.
###Therefore, here I loop through all alignment positions and only keep those above a certain coverage threshold.
##A threshold of 90% means 90% of all proteins aligned don't have a gap at that position.

highcovarray=[]
threshold=0.9
#####################Identify which positions from alignment have XX% coverage##############
LengthofAlign=len(Protein_dict[Protein_name])
totalcounts=[0]*len(Protein_dict[Protein_name])
for i in Protein_dict:
    '''print(len(Protein_dict[i]))'''
    innerloop=0
    for j in Protein_dict[i]:
        if j != '.':
            tempcount=totalcounts[innerloop]+1
            totalcounts[innerloop]=tempcount
        innerloop=innerloop+1
#print('totalcounts',totalcounts)
maxnum=len(Protein_dict.keys())
loopcount10=0
for k in totalcounts:
    if k >= threshold*maxnum:
        '''print(loopcount10)'''
        try:
            highcovarray.append(loopcount10)
        except:
            highcovarray=[loopcount10,]
    loopcount10=1+loopcount10
    
outfile=open('../Results/HighCoverageArray.txt','w')
for i in highcovarray:
    outfile.write('%s,'%(i))
    
#print('highcovarray',highcovarray)
#print(len(highcovarray))


```

```python
def nuc_trans(row):
    if row['Tumor_Seq_Allele1']!=row['Tumor_Seq_Allele2']:
        if row['Reference_Allele']!=row['Tumor_Seq_Allele1']:
            change_value='%s_%s'%(row['Reference_Allele'],row['Tumor_Seq_Allele1'])
        else:
            change_value='%s_%s'%(row['Reference_Allele'],row['Tumor_Seq_Allele2'])
    else:
        #Ignoring mutations to both strands
        print('Both alleles mutated????')
    return(change_value)



```

```python
#########Counting most frequent positions from TCGA for each cancer type#####################





#cancerlist=open('../DATA/Mutations/filelist.txt')#external file with a list of all cancer directories
cancerlist=['SKCM']#Or just list the ones you care about
bins=np.linspace(10,100,10)
for cancer in cancerlist:
    cancertype=cancer

    ###Create RNA expression dictionary for all GPCR's to verify they are actually expresssed in tissues###############
    #Parsing clinical files
    Datadict=parse_clinical(cancertype)
    print('clinical done')
    ##Parsing RNA seq files
    RNAcountdict=Parse_RNA(cancertype,GPCR_protlist)
    #print(RNAcountdict)
    print('rna done')
    
    
    
    
    
    ###Time to finally parse all the mutations and connect to alignment positions
    GPCR_dict={}
    GPCR_array=[]
    mutationarray=[]
    path='%s/%s/'%(mutation_data_path,cancertype)
    checkuniquedict=defaultdict(list)
    for filename in glob.glob(os.path.join(path,'*EA.maf')):
        print(filename)
        maf_file=pd.read_csv(filename,sep='\t')
        
        
        
        ###Before we move on, first going to count the nucleotide change frequencies for 1) All proteins and 
        # 2) only GPCRs. These values will be used later for random controls, just efficient to calculate them now.
        maf_file_nuc=maf_file.loc[:,['Tumor_Sample_Barcode','NM','GENE','SUB','ACTION','Reference_Allele','Tumor_Seq_Allele1','Tumor_Seq_Allele2']]
        maf_file_nuc.drop_duplicates(inplace=True)
        maf_file_nuc['change_value']=maf_file_nuc.apply(nuc_trans,axis=1)

        nuclist=['A','T','C','G','-']
        nuc_df = pd.DataFrame(0, index=nuclist, columns=nuclist)
        for i in maf_file_nuc['change_value']:
            nuc_parts=i.split('_')
            try:
                nuc_df[nuc_parts[0]][nuc_parts[1]]+=1
            except:
                #exceptions to single nuc polymorphisms happen very rarely
                ''
                #print(nuc_parts)
        nuc_df.columns=['A','U','C','G','-']
        nuc_df.index=['A','U','C','G','-']
        
        
        #nucleotide change rates for only GPCRs now
        maf_file_gpcr=maf_file[maf_file['GENE'].isin(GPCR_protlist)]
        maf_file_nuc_gpcr=maf_file_gpcr.loc[:,['Tumor_Sample_Barcode','NM','GENE','SUB','ACTION','Reference_Allele','Tumor_Seq_Allele1','Tumor_Seq_Allele2']]
        maf_file_nuc_gpcr.drop_duplicates(inplace=True)
        maf_file_nuc_gpcr['change_value']=maf_file_nuc_gpcr.apply(nuc_trans,axis=1)

        nuclist=['A','T','C','G','-']
        nuc_gpcr_df = pd.DataFrame(0, index=nuclist, columns=nuclist)
        for i in maf_file_nuc_gpcr['change_value']:
            nuc_parts=i.split('_')
            try:
                nuc_gpcr_df[nuc_parts[0]][nuc_parts[1]]+=1
            except:
                #exceptions to single nuc polymorphisms happen very rarely
                ''
                #print(nuc_parts)
        nuc_gpcr_df.columns=['A','U','C','G','-']
        nuc_gpcr_df.index=['A','U','C','G','-']
        
        
        
        ###Now back to combining mutations
        maf_file_gpcr_df=maf_file_gpcr.loc[:,['Tumor_Sample_Barcode','NM','GENE','SUB','ACTION']]
        maf_file_gpcr_df['tcga_barcode']=maf_file_gpcr_df['Tumor_Sample_Barcode'].str[0:16]
        maf_file_gpcr_df.drop_duplicates(inplace=True)
        maf_file_gpcr_df['rna_value']=maf_file_gpcr_df.apply(check_rna,axis=1)
        maf_file_gpcr_df.reset_index(inplace=True)
        
        
        
        
        
        
        
        ##Alright, so for every line in the mutation file we check to see:
        ##if the mutated position is in the high coverage positions from earlier
        for row in range (0,len(maf_file_gpcr_df)):
            PatientBC=maf_file_gpcr_df.loc[row]['tcga_barcode']
            NMid_mutlist=maf_file_gpcr_df.loc[row]['NM']
            Protein=maf_file_gpcr_df.loc[row]['GENE']
            Mutationlist=maf_file_gpcr_df.loc[row]['SUB']
            EAlist=maf_file_gpcr_df.loc[row]['ACTION']
            rnavalue=maf_file_gpcr_df.loc[row]['rna_value']
        
            
            ###A single substitution may have multiple isoforms.
            ###Options to take highest EA, all EA, first, last, etc
            indNM=NMid_mutlist.split(';')
            indMutation=Mutationlist.split(';')
            indEA=EAlist.split(';')
            NM_ind=indNM[0]
            Mutation_ind=indMutation[0]
            EA_ind=indEA[0]
            
            position=Mutation_ind[1:-1]
            

            ##For each mutation, I am using dictionaries made in past steps in order to convert residue pos
            ##to alignment position in order to group mutations
            testvalue="FALSE"
            try:
                Uniprotname=genename_dict[NM_ind][2]
                truepositionarray=Protein_dict[Uniprotname]
                trueposition=truepositionarray.index(int(position))
                #print(cancertype,Uniprotname,Mutation_ind,trueposition)
                testvalue="TRUE"
            except:
                testvalue="FALSE"
                try:
                    errorlog.append(NM_ind)
                except:
                    errorlog=NM_ind
                    

            if testvalue=="TRUE" and trueposition in highcovarray:

                temp_tuple=trueposition,Protein,Mutation_ind,EA_ind,NM_ind,PatientBC,rnavalue

                #print("success",temp_tuple)
                try:
                    GPCR_array.append(temp_tuple)
                except:
                    GPCR_array=temp_tuple

    #print(checkuniquedict)
    
    mutation_df=pd.DataFrame(GPCR_array,columns=['align_pos','Gene','Mutation','EA_score','NM_id','Patient_ID','RNA_seq'])
    saving_files(mutation_df,cancertype,nuc_df,nuc_gpcr_df)
    print('done with %s'%(cancertype))
```
